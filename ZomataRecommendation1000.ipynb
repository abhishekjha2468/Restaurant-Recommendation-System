```python
import pandas as pd

zf = pd.read_csv('zomato.csv')
```

```python
# To seperate all the rate and reviews from the reviews column from a single restaurant
# because there are multiple reviews for a single restaurant

import re
from tqdm import tqdm
rate = []
reviews = []
restName = zf['name']
loc = zf['reviews_list']
name = []
f=0

for idx, row in tqdm(enumerate(zip(loc, restName))):
  i, r = row
  temp = list(re.findall("\((.*?)\)", i))
  for j in temp:
    if "Rated" not in j or "RATED" not in j:
      continue
    try:
      try:
        a, b = (j).split("RATED\\n")
      except:
        a, b = (j).split("RATED\n")
      
      rate.append(float(a.split("Rated")[1][:3]))
      
      reviews.append(b[:len(b)-1])
      name.append(r)
    except Exception as e:
      print(e, idx, j)
      print((j.split("', 'RATED\\n ")))
      f=1
      break
  if f==1:
    break
```

```python
data = pd.DataFrame()

data['reviews'] = reviews
data['name'] = name
data['rate'] = rate
```

```python
data.to_csv('/content/drive/MyDrive/pureZomato.csv')
```

```python
# Cleaning the reviews
import nltk
from nltk.corpus import stopwords 
nltk.download('stopwords')
import string
import re
def clean_text(text):
    text = text.translate(string.punctuation)
    
    ## Convert words to lower case and split them
    text = text.lower().split()
    
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops and len(w) >= 3]
    
    text = " ".join(text)
    
    # Clean the text
    text = re.sub(r"[^A-Za-z0-9^,!.\/'+-=]", " ", text)
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r",", " ", text)
    text = re.sub(r"\.", " ", text) 
    text = re.sub(r"!", " ", text)
    text = re.sub("(\s\d+)","",text)
    text = re.sub(r":", " ", text)
    text = re.sub(r"\s\S*\d{1,}\S*\s", " ", text)
    text = re.sub(r"\/", " ", text)
    text = re.sub(r"\^", " ", text)
    text = re.sub(r"\+", " ", text)
    text = re.sub(r"\-", " ", text)
    text = re.sub(r"\=", " ", text)
    text = re.sub(r"'", " ", text)
    text = re.sub(r"(\d+)(k)", r"\g<1>000", text)
    text = re.sub(r":", " : ", text)
    text = re.sub(r" e g ", " eg ", text)
    text = re.sub(r" b g ", " bg ", text)
    text = re.sub(r" u s ", " american ", text)
    text = re.sub(r"\0s", "0", text)
    text = re.sub(r" 9 11 ", "911", text)
    text = re.sub(r"e - mail", "email", text)
    text = re.sub(r"j k", "jk", text)
    text = re.sub(r"\s{2,}", " ", text)
    #removing symbols
    text = re.sub(r"(\d+|\W+)", ' ', text)
    #data[fea]=data[fea].str.replace(r"(\d+|\W+)", ' ',regex=True)
    print("SYMBOLS REMOVED")
    #removing 2 letters words
    text = re.sub(r"\b\w{1,2}\b", '', text)
    #data[fea]=data[fea].str.replace(r"\b\w{1,2}\b", '',regex=True)
    print("2 letter words REMOVED")
    #removing extra spaces
    for i in range(4):
      text = re.sub(" "*2, ' ', text)
    print("EXTRA SPACES REMOVED")    
    return text
```

# Matrix Factorization

```python
# Creating matrix of restaurant name and reviews. The matrix will contain the values 
# of rating
rating_matrix = pd.pivot_table(abc, values='rate', index=['name'], columns=['review'])
rating_matrix.shape
```

```python
rating_matrix.head()
```

```python
# Here we are combining reviews of each unique reastaurant name.
newDf = abc.groupby('name').agg({'review': ' '.join})
```

```python
newDf.shape
```

```python
# We are converting the unique reviews into vectors using tfidf 

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import WordPunctTokenizer


review_vectorizer = TfidfVectorizer(tokenizer = WordPunctTokenizer().tokenize, max_features=1000)
review_vectors = review_vectorizer.fit_transform(abc['review'].unique())
review_vectors.shape
```

```python
# We are converting the combined reviews of unique restaurant name into vectors using tfidf 

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import WordPunctTokenizer

name_vectorizer = TfidfVectorizer(tokenizer = WordPunctTokenizer().tokenize, max_features=1000)
name_vectors = name_vectorizer.fit_transform(pd.DataFrame(newDf)['review'])
name_vectors.shape
```

```python
# Creating two Matrix P and Q 
# P contains vector corresponding to each restaurant name
# Q contains vector corresponding to each review

P = pd.DataFrame(name_vectors.toarray(), index=newDf.index, columns=name_vectorizer.get_feature_names())
Q = pd.DataFrame(review_vectors.toarray(), index=abc.review.unique(), columns=review_vectorizer.get_feature_names())
```

```python
P.head(2)
```

```python
# We are using Gradient descent to predict the missing matrix values

def matrix_factorization(R, P, Q, steps=25, gamma=0.001,lamda=0.02):
    for step in range(steps):
        for i in R.index:
            for j in R.columns:
                if R.loc[i,j]>0:
                    eij=R.loc[i,j]-np.dot(P.loc[i],Q.loc[j])
                    P.loc[i]=P.loc[i]+gamma*(eij*Q.loc[j]-lamda*P.loc[i])
                    Q.loc[j]=Q.loc[j]+gamma*(eij*P.loc[i]-lamda*Q.loc[j])
        e=0
        for i in R.index:
            for j in R.columns:
                if R.loc[i,j]>0:
                    e= e + pow(R.loc[i,j]-np.dot(P.loc[i],Q.loc[j]),2)+lamda*(pow(np.linalg.norm(P.loc[i]),2)+pow(np.linalg.norm(Q.loc[j]),2))
        if e<0.001:
            break
        
    return P,Q
```

```python
%%time
p, q = matrix_factorization(rating_matrix, P, Q, steps=25, gamma=0.001,lamda=0.02)
```

```python
# Store P, Q and vectorizer in pickle file
import pickle
output = open('/content/drive/MyDrive/zomato1000_recommendation_model.pkl', 'wb')
pickle.dump(p,output)
pickle.dump(q,output)
pickle.dump(name_vectorizer,output)
output.close()
```

```python
# Prediction for input word and store the result in topRecommendations
words = "hygienic"
test_df= pd.DataFrame([words], columns=['review'])
test_df['review'] = test_df['review'].apply(clean_text)
test_vectors = review_vectorizer.transform(test_df['review'])
test_v_df = pd.DataFrame(test_vectors.toarray(), index=test_df.index, columns=review_vectorizer.get_feature_names())

predictItemRating=pd.DataFrame(np.dot(test_v_df.loc[0],q.T),index=q.index,columns=['rate'])
topRecommendations=pd.DataFrame.sort_values(predictItemRating,['rate'],ascending=[0])[:7]
```
